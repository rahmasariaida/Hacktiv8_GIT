# Unsupervised Learning

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breastData = breast.data
```

```python
breastData.shape
```

```python
breastLabels = breast.target
```

```python
breastLabels.shape
```

```python
import numpy as np
```

```python
labels = np.reshape(breastLabels,(569,1))
```

```python
finalBreastData = np.concatenate([breastData, labels], axis=1)
```

```python
finalBreastData.shape
```

```python
import pandas as pd
```

```python
breastDataset = pd.DataFrame(finalBreastData)
```

```python
breastDataset.head(3)
```

```python
features = breast.feature_names
```

```python
features
```

```python
featuresLabels = np.append(features, 'label')
```

```python
breastDataset.columns = featuresLabels
```

```python
breastDataset.head(3)
```

```python
breastDataset['label'].replace(0, 'Benign', inplace=True)
breastDataset['label'].replace(1, 'Malignant', inplace = True)
```

```python
breastDataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=breastDataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
x
```

```python
normalisedBreast = pd.DataFrame(x)
normalisedBreast.tail()
```

```python
from sklearn.decomposition import PCA

pcaBreast=PCA(n_components=2)
principalComp_Breast=pcaBreast.fit_transform(x)
```

```python
pcBreastDF = pd.DataFrame(data=principalComp_Breast,
                         columns= [
                             'principal component 1',
                             'principal component 2'
                         ])
```

```python
pcBreastDF.head()
```

```python
pcaBreast.explained_variance_ratio_
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title('Principal Component Analysis of Breast Cancer Dataset', fontsize =20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets, colors):
    indicesToKeep = breastDataset['label'] == target
    plt.scatter(pcBreastDF.loc[indicesToKeep, 'principal component 1']
                , pcBreastDF.loc[indicesToKeep, 'principal component 2'], c = color, s =50)

plt.legend(targets, prop={'size':15})

plt.show()
```

## K Means CLustering

```python
import pandas as np
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataScaled = scaler.fit_transform(data)
```

```python
pd.DataFrame(dataScaled).describe()
```

```python
kmeans=KMeans(n_clusters=2, init='k-means++') #max_iter

kmeans.fit(dataScaled)
```

```python
kmeans.inertia_
```

```python
SSE =[] #Sum Square Error
for cluster in range(1,20):
    kmeans=KMeans(n_jobs=-1, n_clusters=cluster, init= 'k-means++')
    kmeans.fit(dataScaled)
    SSE.append(kmeans.inertia_)

frame= pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans=KMeans(n_clusters=6, init= 'k-means++')
kmeans.fit(dataScaled)
```

```python
pred=kmeans.predict(dataScaled)
```

```python
frame =pd.DataFrame(dataScaled)
frame['cluster'] = pred
```

```python
frame
```
